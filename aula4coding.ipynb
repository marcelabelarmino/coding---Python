```python
cupom = input ('Insira o cupom:')
if (cupom == 'aula1'):
    print ('Você ganhou 15% de desconto!')
elif (cupom == 'aula2'):
    print ('Você recebeu 10% de desconto!')
else:
    print ('Você recebeu 5% de desconto!')
```

```python
sol = input ('Está fazendo sol?, se sim, digite 1, senão digite 2?')
dinheiro = input ('Você tem dinheiro, se sim, digite 1, senão digite 2?')

if (sol and dinheiro == '1'):
    print ('Vá à praia!')
elif (sol and dinheiro != '1'):
    print ('Dia de Netflix!')
```

```python
sol = int(input ('Está fazendo sol?, se sim, digite 1, senão digite 2?'))
if (sol == '1'):
    dinheiro = input ('Você tem dinheiro?, se sim, digite 1, senão digite 2?')
if (sol and dinheiro == '1'):
    print ('Vá a praia!')
else:
    print ('Veja Netflix')
```
